# MatPlot

```python
import matplotlib.pyplot as plt
```

```python
# les données a visualiser
prix_ms= [39, 50, 57, 69, 94, 100]
prix_apple=[93,112,104,144,170, 172]
annee=[2015, 2016,2017,2018,2019, 2020]
```

```python
#visualiser les données
plt.scatter(annee, prix_ms, label='Prix windows', c='g')
plt.plot(annee, prix_apple, label='Prix Apple', c='y')
plt.legend()
plt.xlabel('Année')
plt.ylabel('Prix des ordinateurs')
plt.title('Prix des ordinateurs suivant les années')
```

```python
#visualiser les données
plt.subplot(2,1,1)
plt.scatter(annee, prix_ms, label='Prix windows', c='g')

plt.subplot(2,1,2)
plt.plot(annee, prix_apple, label='Prix Apple', c='y')
plt.legend()
plt.xlabel('Année')
plt.ylabel('Prix des ordinateurs')
plt.title('Prix des ordinateurs suivant les années')
```

```python
#visualiser les données en diagramme de bandee
import numpy as np
index_annee = np.arange(len(annee))
plt.bar(index_annee, prix_apple, label='prix apple', width=0.25)
plt.bar(index_annee+0.25, prix_ms, label='prix microsoft', width=0.25)
plt.legend()
plt.xlabel("Annee")
plt.ylabel("prix des apple")
plt.title("Evolution des prix des ordinateurs durant les 5 Dernieres Annees")
```
